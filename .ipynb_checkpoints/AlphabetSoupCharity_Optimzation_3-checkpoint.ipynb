### Deliverable 1: Preprocessing the Data for a Neural Network

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,OneHotEncoder
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd 
application_df = pd.read_csv("Resources/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
application_df = application_df.drop('NAME', axis=1)
application_df = application_df.drop('EIN', axis=1)
application_df.head()
```

```python
# Determine the number of unique values in each column.
print(application_df.nunique())
```

```python
# Look at APPLICATION_TYPE value counts for binning
app_counts = application_df.value_counts('APPLICATION_TYPE')
app_counts
```

```python
# Visualize the value counts of APPLICATION_TYPE
app_counts.plot.density()
```

```python
# Determine which values to replace if counts are less than ...?
replace_app = list(app_counts[app_counts < 500].index)

# Replace in dataframe
for app in replace_app:
    application_df.APPLICATION_TYPE = application_df.APPLICATION_TYPE.replace(app,"Other")
    
# Check to make sure binning was successful
application_df.APPLICATION_TYPE.value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
class_counts = application_df.value_counts('CLASSIFICATION')
class_counts
```

```python
# Visualize the value counts of CLASSIFICATION
class_counts.plot.density()
```

```python
# Determine which values to replace if counts are less than ..?
replace_class = list(class_counts[class_counts < 1000].index)

# Replace in dataframe
for cls in replace_class:
    application_df.CLASSIFICATION = application_df.CLASSIFICATION.replace(cls,"Other")
    
# Check to make sure binning was successful
application_df.CLASSIFICATION.value_counts()
```

```python
# Generate our categorical variable lists
application_cat = application_df.dtypes[application_df.dtypes == "object"].index.tolist()
application_cat
```

```python
# Create a OneHotEncoder instance
enc = OneHotEncoder(sparse=False)

# Fit and transform the OneHotEncoder using the categorical variable list
encode_df = pd.DataFrame(enc.fit_transform(application_df[application_cat]))

# Add the encoded variable names to the dataframe
encode_df.columns = enc.get_feature_names(application_cat)
encode_df.head()
```

```python
# Merge one-hot encoded features and drop the originals
application_df = application_df.merge(encode_df,left_index=True, right_index=True)
application_df = application_df.drop(application_cat,1)
application_df.head()
```

```python
# Split our preprocessed data into our features and target arrays
y = application_df['IS_SUCCESSFUL'].values
X = application_df.drop(['IS_SUCCESSFUL'],1).values

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

### Deliverable 2: Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.

# In this optimzation, chanhig number outputs, hidden nodes, adding a hidden node layer and using leakyrelu

nn = tf.keras.models.Sequential()
number_input_features = len(X_train[0])
number_inputs = 100
number_hidden_nodes1 = 30
number_hidden_nodes2 = 20

# First hidden layer
nn.add(
    tf.keras.layers.Dense(units=number_inputs, input_dim=number_input_features, activation=tf.keras.layers.LeakyReLU())
)


# Second hidden layer
nn.add(tf.keras.layers.Dense(units=number_hidden_nodes1, activation=tf.keras.layers.LeakyReLU()))

# Third hidden layer
nn.add(tf.keras.layers.Dense(units=number_hidden_nodes1, activation=tf.keras.layers.LeakyReLU()))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Import checkpoint dependencies
import os
from tensorflow.keras.callbacks import ModelCheckpoint
```

```python
# Callback to save model every 5 epochs

# Define the checkpoint path and filenames
os.makedirs("checkpoints/", exist_ok=True)
checkpoint_path = "checkpoints/weights.{epoch:02d}hdf5"


cp_callback = ModelCheckpoint(
    filepath=checkpoint_path,
    verbose=1,
    save_weights_only=True,
    save_freq='epoch',
    period=5)
```

```python
# Train the model
fit_model = nn.fit(X_train,y_train,epochs=250)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# Export our model to HDF5 file
nn.save("AlphabetSoupCharity_Optimization_3.h5")
```
